```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
t=np.linspace(-1,2.5,1000)
def cos(Function,x):
    Y=np.cos(Function(x))
    return Y

def plot(x,y,color,fun):
    plt.title('Plot of diffrent cos Functions')
    plt.plot(x,y,color=color)
    plt.xlabel('Time')
    plt.ylabel('Y value')
    print(color,'=',fun)
    
one  =lambda t:2*np.pi*t
two  =lambda t:2*np.pi*t+np.pi
three=lambda t:8*np.pi*t
four =lambda t:4*np.pi*t-(np.pi/3)

Y1=cos(one,t)
Y2=cos(two,t)
Y3=cos(three,t)
Y4=cos(four,t)
y=[[Y1,'red','2*np.pi*t'],[Y2,'green','2*np.pi*t+np.pi'],[Y3,'blue','8*np.pi*t'],[Y4,'orange','4*np.pi*t-(np.pi/3)']]
for Y,color,fun in (y):
    mpl.style.use('seaborn')
    plot(t,Y,color,fun)

    
plt.show()
```

```python
x=np.linspace(-1.5,2.5,1000,endpoint=True)
y=np.cos(x+0.25)
y2=np.cos(2*np.pi*(x+0.25))-0.6
mpl.style.use('seaborn')
plt.plot(x,y)
plt.plot(x,y2)
```
