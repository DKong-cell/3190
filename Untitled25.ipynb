```python
import numpy as np
import matplotlib.pyplot as plt

k=8
m=2
A=0.4
phi=-0.4
omega=np.sqrt(k/m)


t=np.linspace(0,2*np.pi,1000)
x=A*np.cos(omega*t+phi)
v=-omega*A*(np.sin(omega*t+phi))
p=m*v


plt.grid(True)
plt.plot(x,p)
plt.show()

"""


"""
```
